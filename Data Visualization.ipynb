```python
from scipy.stats import chi2_contingency
```

```python
# Create a contingency table for two categorical variables
contingency_table = pd.crosstab(df['Electric Vehicle Type'], df['Make'])
```

```python
# Display the contingency table
print("Contingency Table:")
print(contingency_table)
```

```python
# Perform chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)
```

```python
# Display test statistics and p-value
print("\nChi-square Statistic:", chi2)
print("P-value:", p)
```

```python
# Interpret the results
alpha = 0.05
print("\nSignificance Level (alpha):", alpha)

if p < alpha:
    print("Reject the null hypothesis: There is a significant association between the categorical variables.")
else:
    print("Fail to reject the null hypothesis: There is no significant association between the categorical variables.")
```

```python
from scipy.stats import ttest_ind
```

```python
# Display basic information about the DataFrame
print("DataFrame Info:")
print(df.info())
```

```python
# Display summary statistics for each group
group1_stats = df[df['Make'] == 'Electric Range']['Model Year'].describe()
group2_stats = df[df['Electric Vehicle Type'] == 'Model']['Model Year'].describe()
```

```python
print("\nGroup 1 Statistics:")
print(group1_stats)
```

```python
print("\nGroup 2 Statistics:")
print(group2_stats)
```

```python
# Perform independent t-test
t_stat, p_value = ttest_ind(df[df['Make'] == 'Electric Range']['Model Year'],
                            df[df['Electric Vehicle Type'] == 'Model']['Model Year'])
```

```python
print("\nIndependent t-test:")
print("t-statistic:", t_stat)
print("p-value:", p_value)
```
